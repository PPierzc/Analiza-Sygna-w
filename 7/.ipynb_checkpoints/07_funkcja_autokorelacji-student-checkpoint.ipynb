# Funkcja autokorelacji
## Funkcja korelacji wzajemnej

Dla sygnału  $x(t)$  możemy badać jak jest on do siebie podobny dla przesunięciaczasowego $\tau $:

  $\sigma _{xx}(\tau ) = \int {x(t) x(t+\tau ) dt} $

Zwróćmy uwagę na związek funkcji korelacji:
* z iloczynem skalarnym wektorów $x(t)$ i jego przesuniętej wersji $x(t +\tau)$
* ze splotem.

Dla skończonych dyskretnych sygnałów mamy estymatory korelacji:
  $R_{xx}(m) = E\lbrace x_{n+m}x^*_n \rbrace  = E\lbrace x_{n}x^*_{n-m} \rbrace $

oraz kowariancji:
  $C_{xx}(m) = E\lbrace (x_{n+m}-\bar{x}) (x_n-\bar{x})^*\rbrace  = R_{xx}(m) - \bar{x} \bar{x}^*$

Funkcję $R$ można estymować z jednej realizacji procesu (zakładamy jego ergodyczność):
  $ \widehat{R}_{xx}(m) = \left\lbrace 
\begin{array}{ll}
\sum _{n=0}^{N-m-1}{x_{n+m} x_n^*} & m \ge 0 \\
\widehat{R}_{xx}^*(-m) & m < 0
\end{array} \right.$

### Zadanie: Jak obliczyć funkcję autokorelacji?
W pythonie mamy wydajną implementację funkcji autokorelacji (tak naprawdę to funkcji korelacji wzajemnej, której szczególnym przypadkiem jest autokorelacja)
w: [numpy.correlate](http://docs.scipy.org/doc/numpy/reference/generated/numpy.correlate.html ). Zanim jednak przejdziemy do rutynowego korzystania z tej funkcji warto sobie wyrobić pewną intuicję co do jej działania. Możemy sobie to działanie wyobrazić tak, że tworzymy dwie kopie naszego sygnału wydłużone zerami. jedna kopia ma oryginalny sygnał po środku, a w drugiej będzie on stopniowo przesuwany od pozycji takiej w której z nieruchomą kopią ma jeden punkt wspólny z lewej stron do takiej, kiedy ma jeden punkt wspólny z prawej strony. W każdym położeniu wzajemnym sygnałów musimy policzyć ich iloczyn skalarny.  Poniższy kod pomoże nam to zilustrować:

```python
# Niech nasz sygnał będzie:
x = np.array([1,2,3,2,1])

# Ma on długość
N = len(x)

# przygotujmy miejsce na kopie sygnału
# najpierw kopia nieruchoma -> nasz sygnał jest na środku
x_2 = np.zeros(3*N-2, dtype = 'int')
x_2[N-1: 2*N-1] = x
print('liczymy f. autokorelacji sygnału x= ',x)
f_corr = np.zeros(2*N)
for k in range(-N,N-1):
    x_1 = np.zeros(3*N-2, dtype = 'int')
    x_1[N+k:k+2*N]=x  # to jest kopia sygnału x cofnięta o k próbek w stosunku do nieruchomej kopii
    f_corr[k+N] = np.sum(np.dot(x_1,x_2)) # 
    # robimy wydruk obliczeń:
    print(' ')
    print('x1: \t'+str(x_1))
    print('x2: \t'+str(x_2))
    print(40*'-')
    print('x1*x2\t', x_1*x_2)  
    print('f_corr(',k+1,') = \t', f_corr[k+N])
# porównajmy wynik z wywołaniem funkcji bibliotecznej:
print(' ')
print(51*'_')
print(f_corr) # co zwraca nasza funkcja
print(np.correlate(x,x, 'full') ) # co zwraca f. biblioteczna
```

### Zadanie: Funkcja auokorelacji funkcji okresowej
Funkcja okresowa jest do siebie podobna po przesunięciu o cały okres. 
* Jak zatem wygląda funkcja autokorelacji fragmentu funkcji sinus?
* Jak zmienia się ta funkcja wraz z długością fragmentu sygnału?

```python
Fs = 100
dt = 1/Fs
f = 10
T = 0.2
t = np.arange(0,T,dt)
s = np.sin(2*np.pi*f*t)
N = len(s)

norm_f = np.hstack((np.arange(1,N,1), np.arange(N,0,-1)))

f_corr = np.correlate(s,s,mode='full') # pełna wersja funkcji autokorelacji - za pomoca funkcji bibliotecznej
tau  = np.arange(-N+1,N, 1) # wektor przesunięć wzajemnych - powinien mieć 0 na środku

plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.stem(t,s)

plt.subplot(2,1,2)
plt.stem(tau,f_corr/norm_f)

plt.show()
```

### Zadanie: Funkcja auokorelacji białego szumu
Analogicznie do zadania powyżej proszę zapoznać się z funknją autokorelacji dla białego szumu. Dla ustalenia uwagi niech to będą niezależne próbki z rozkładu N(0,1).

```python
import scipy.stats as st
Fs = 10000
dt = 1/Fs
f = 10
T = 0.2
t = np.arange(0,T,dt)
s = st.norm.rvs(loc=0, scale=1, size = len(t))
N = len(s)

f_corr = np.correlate(s,s,mode='full') # pełna wersja funkcji autokorelacji - za pomoca funkcji bibliotecznej
tau  = np.arange(-N+1,N, 1) # wektor przesunięć wzajemnych - powinien mieć 0 na środku

plt.figure(figsize=(20,10))
plt.subplot(2,1,1)
plt.plot(t,s)

plt.subplot(2,1,2)
plt.plot(tau,f_corr)

plt.show()
```
